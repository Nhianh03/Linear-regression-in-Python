Gi·ªõi thi·ªáu v·ªÅ **H·ªìi quy tuy·∫øn t√≠nh (Linear Regression)**

**H·ªìi quy tuy·∫øn t√≠nh (Linear Regression)** l√† m·ªôt trong nh·ªØng thu·∫≠t to√°n c∆° b·∫£n v√† ph·ªï bi·∫øn nh·∫•t trong h·ªçc c√≥ **gi√°m s√°t** (Supervised Learning). N√≥ ch·ªß y·∫øu ƒë∆∞·ª£c s·ª≠ d·ª•ng cho b√†i **to√°n h·ªìi quy**, t·ª©c l√† d·ª± ƒëo√°n m·ªôt gi√° tr·ªã li√™n t·ª•c.

**Nguy√™n l√Ω ho·∫°t ƒë·ªông c·ªßa Linear Regression**

**Linear Regression** t√¨m c√°ch x√°c ƒë·ªãnh m·ªëi quan h·ªá tuy·∫øn t√≠nh gi·ªØa bi·∫øn ƒë·∫ßu v√†o (ƒë·ªôc l·∫≠p, k√Ω hi·ªáu l√† X) v√† bi·∫øn ƒë·∫ßu ra (ph·ª• thu·ªôc, k√Ω hi·ªáu l√† Y). Thu·∫≠t to√°n s·∫Ω c·ªë g·∫Øng t√¨m m·ªôt ƒë∆∞·ªùng th·∫≥ng t·ªët nh·∫•t ƒë·ªÉ m√¥ t·∫£ m·ªëi quan h·ªá n√†y.


‚úÖ Thu th·∫≠p d·ªØ li·ªáu: L·∫•y d·ªØ li·ªáu ƒë·∫ßu v√†o (X) v√† ƒë·∫ßu ra (Y).
‚úÖ Ti·ªÅn x·ª≠ l√Ω d·ªØ li·ªáu: X·ª≠ l√Ω d·ªØ li·ªáu b·ªã thi·∫øu, chu·∫©n h√≥a d·ªØ li·ªáu n·∫øu c·∫ßn.
‚úÖ X√¢y d·ª±ng m√¥ h√¨nh: T√¨m ƒë∆∞·ªùng th·∫≥ng t·ªëi ∆∞u c√≥ d·∫°ng:

Ph∆∞∆°ng tr√¨nh c·ªßa ƒë∆∞·ªùng h·ªìi quy c√≥ d·∫°ng:

Y = mX + b

trong ƒë√≥:
	‚Ä¢	 Y  l√† gi√° tr·ªã c·∫ßn d·ª± ƒëo√°n

	‚Ä¢	 X  l√† bi·∫øn ƒë·ªôc l·∫≠p
	
	‚Ä¢	 m  l√† h·ªá s·ªë g√≥c (Slope)
	
	‚Ä¢	 b  l√† h·ªá s·ªë ch·∫∑n (Intercept)

‚úÖ Hu·∫•n luy·ªán m√¥ h√¨nh: S·ª≠ d·ª•ng thu·∫≠t to√°n t·ªëi ∆∞u (nh∆∞ Gradient Descent) ƒë·ªÉ t√¨m gi√° tr·ªã  m  v√†  b .

‚úÖ ƒê√°nh gi√° m√¥ h√¨nh: Ki·ªÉm tra ƒë·ªô ch√≠nh x√°c b·∫±ng c√°c ch·ªâ s·ªë nh∆∞ R¬≤ Score, Mean Squared Error (MSE).

‚úÖ D·ª± ƒëo√°n: Sau khi m√¥ h√¨nh ƒë∆∞·ª£c hu·∫•n luy·ªán, s·ª≠ d·ª•ng n√≥ ƒë·ªÉ ƒë∆∞a ra d·ª± ƒëo√°n tr√™n d·ªØ li·ªáu m·ªõi.

# **3.  How to find the best straight line?** <a class="anchor" id="3"></a>

[Table of Contents](#0.1)


H·ªìi quy tuy·∫øn t√≠nh s·ª≠ d·ª•ng ph∆∞∆°ng ph√°p **b√¨nh ph∆∞∆°ng sai s·ªë nh·ªè nh·∫•t** (Ordinary Least Squares - OLS) ƒë·ªÉ t√¨m ƒë∆∞·ªùng th·∫≥ng t·ªëi ∆∞u. M·ª•c ti√™u c·ªßa ph∆∞∆°ng ph√°p n√†y l√† **t·ªëi thi·ªÉu h√≥a t·ªïng b√¨nh ph∆∞∆°ng** kho·∫£ng c√°ch t·ª´ c√°c ƒëi·ªÉm d·ªØ li·ªáu th·ª±c t·∫ø ƒë·∫øn ƒë∆∞·ªùng th·∫≥ng d·ª± ƒëo√°n.


\text{Loss Function} = \sum (Y_{\text{th·ª±c t·∫ø}} - Y_{\text{d·ª± ƒëo√°n}})^2

C√†ng √≠t sai s·ªë, ƒë∆∞·ªùng th·∫≥ng c√†ng ph√π h·ª£p v·ªõi d·ªØ li·ªáu.

üìå Tr·ª±c gi√°c v·ªÅ ƒë·ªô d·ªëc  m  v√† h·ªá s·ªë ch·∫∑n  b 
	
    ‚Ä¢	N·∫øu  m > 0 : ƒê∆∞·ªùng th·∫≥ng c√≥ xu h∆∞·ªõng ƒëi l√™n, nghƒ©a l√† khi  X  tƒÉng,  Y  c≈©ng tƒÉng (v√≠ d·ª•: di·ªán t√≠ch nh√† l·ªõn h∆°n ‚Üí gi√° nh√† cao h∆°n).
	
    ‚Ä¢	N·∫øu  m < 0 : ƒê∆∞·ªùng th·∫≥ng ƒëi xu·ªëng, nghƒ©a l√† khi  X  tƒÉng,  Y  gi·∫£m (v√≠ d·ª•: tƒÉng tu·ªïi c·ªßa m·ªôt chi·∫øc xe ‚Üí gi√° tr·ªã xe gi·∫£m).
	
    ‚Ä¢	H·ªá s·ªë ch·∫∑n  b  th·ªÉ hi·ªán gi√° tr·ªã  Y  khi  X = 0 .

üìå V√≠ d·ª• th·ª±c t·∫ø

B·∫°n c√≥ m·ªôt b·ªô d·ªØ li·ªáu ch·ª©a gi√° nh√† ·ªü H√† N·ªôi theo di·ªán t√≠ch (m¬≤). N·∫øu √°p d·ª•ng h·ªìi quy tuy·∫øn t√≠nh, b·∫°n c√≥ th·ªÉ nh·∫≠n ƒë∆∞·ª£c ph∆∞∆°ng tr√¨nh:

\text{Gi√° nh√†} = 50 \times \text{Di·ªán t√≠ch} + 200

ƒêi·ªÅu n√†y c√≥ nghƒ©a l√†:
	   
 ‚Ä¢	V·ªõi m·ªói m¬≤ tƒÉng th√™m, gi√° nh√† tƒÉng trung b√¨nh 50 tri·ªáu VND.

 ‚Ä¢	Khi di·ªán t√≠ch b·∫±ng 0, gi√° nh√† kh·ªüi ƒëi·ªÉm l√† 200 tri·ªáu VND.

### Box -Cox transformation
Ph√©p bi·∫øn ƒë·ªïi Box Cox l√† m·ªôt c√°ch ƒë·ªÉ bi·∫øn ƒë·ªïi c√°c bi·∫øn ph·ª• thu·ªôc kh√¥ng b√¨nh th∆∞·ªùng th√†nh d·∫°ng b√¨nh th∆∞·ªùng. Chu·∫©n l√† m·ªôt gi·∫£ ƒë·ªãnh quan tr·ªçng ƒë·ªëi v·ªõi nhi·ªÅu k·ªπ thu·∫≠t th·ªëng k√™; n·∫øu d·ªØ li·ªáu c·ªßa b·∫°n kh√¥ng b√¨nh th∆∞·ªùng, vi·ªác √°p d·ª•ng Box-Cox c√≥ nghƒ©a l√† b·∫°n c√≥ th·ªÉ ch·∫°y nhi·ªÅu ph√©p th·ª≠ h∆°n. T·∫•t c·∫£ nh·ªØng g√¨ ch√∫ng ta c·∫ßn ƒë·ªÉ th·ª±c hi·ªán ph√©p bi·∫øn ƒë·ªïi n√†y l√† t√¨m gi√° tr·ªã lambda v√† √°p d·ª•ng quy t·∫Øc ƒë∆∞·ª£c hi·ªÉn th·ªã b√™n d∆∞·ªõi cho bi·∫øn c·ªßa b·∫°n. 
$$\mathbf{ \begin {cases}\frac {y^\lambda - 1}{\lambda},& y_i\neg=0 \\
 log(y_i) & \lambda = 0 \end{cases}}$$
 Th·ªß thu·∫≠t c·ªßa ph√©p bi·∫øn ƒë·ªïi Box-Cox l√† t√¨m gi√° tr·ªã lambda, tuy nhi√™n tr√™n th·ª±c t·∫ø, ƒëi·ªÅu n√†y kh√° d·ªÖ d√†ng. H√†m sau tr·∫£ v·ªÅ bi·∫øn ƒë√£ bi·∫øn ƒë·ªïi, gi√° tr·ªã lambda, kho·∫£ng tin c·∫≠y

# **10. Split data into separate training and test set** <a class="anchor" id="10"></a>

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X = df.drop('charges',axis=1) # Independet variable
y = df['charges'] # dependent variable

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=23)
```

# **11. BUILD MODEL** <a class="anchor" id="11"></a>

[Table of Contents](#0.1)


Trong b∆∞·ªõc n√†y, x√¢y d·ª±ng m√¥ h√¨nh b·∫±ng ph∆∞∆°ng tr√¨nh h·ªìi quy tuy·∫øn t√≠nh $\mathbf{\theta = (X^T X)^{-1} X^Ty}$. ·ªû b∆∞·ªõc ƒë·∫ßu ti√™n, ch√∫ng ta c·∫ßn th√™m m·ªôt t√≠nh nƒÉng $\mathbf{x_0 =1}$ v√†o t·∫≠p d·ªØ li·ªáu g·ªëc c·ªßa m√¨nh.

```python
# Step 1: add x0 =1 to dataset
X_train_0 = np.c_[np.ones((X_train.shape[0],1)),X_train]
X_test_0 = np.c_[np.ones((X_test.shape[0],1)),X_test]

# Step2: build model
theta = np.matmul(np.linalg.pinv(np.matmul(X_train_0.T, X_train_0)), np.matmul(X_train_0.T, y_train))
```

# **12. The parameters for linear regression modelg** <a class="anchor" id="12"></a>

## ƒê√°nh gi√° m√¥ h√¨nh
Ch√∫ng t√¥i s·∫Ω d·ª± ƒëo√°n gi√° tr·ªã cho bi·∫øn m·ª•c ti√™u b·∫±ng c√°ch s·ª≠ d·ª•ng tham s·ªë m√¥ h√¨nh c·ªßa ch√∫ng t√¥i cho t·∫≠p d·ªØ li·ªáu th·ª≠ nghi·ªám. Sau ƒë√≥, so s√°nh gi√° tr·ªã d·ª± ƒëo√°n v·ªõi gi√° tr·ªã th·ª±c t·∫ø trong t·∫≠p d·ªØ li·ªáu th·ª≠ nghi·ªám. Ch√∫ng t√¥i t√≠nh **L·ªói b√¨nh ph∆∞∆°ng trung b√¨nh** b·∫±ng c√¥ng th·ª©c
$$\mathbf{ J(\theta) = \frac{1}{m} \sum_{i=1}^{m}(\hat{y}_i - y_i)^2}$$

$\mathbf{R^2}$ l√† th∆∞·ªõc ƒëo th·ªëng k√™ v·ªÅ m·ª©c ƒë·ªô g·∫ßn c·ªßa d·ªØ li·ªáu v·ªõi ƒë∆∞·ªùng h·ªìi quy ƒë∆∞·ª£c ƒëi·ªÅu ch·ªânh. $\mathbf{R^2}$ lu√¥n n·∫±m trong kho·∫£ng t·ª´ 0 ƒë·∫øn 100%. 0% ch·ªâ ra r·∫±ng m√¥ h√¨nh kh√¥ng gi·∫£i th√≠ch ƒë∆∞·ª£c b·∫•t k·ª≥ s·ª± thay ƒë·ªïi n√†o c·ªßa d·ªØ li·ªáu ph·∫£n h·ªìi xung quanh gi√° tr·ªã trung b√¨nh c·ªßa n√≥. 100% ch·ªâ ra r·∫±ng m√¥ h√¨nh gi·∫£i th√≠ch ƒë∆∞·ª£c t·∫•t c·∫£ s·ª± thay ƒë·ªïi c·ªßa d·ªØ li·ªáu ph·∫£n h·ªìi xung quanh gi√° tr·ªã trung b√¨nh.

$$\mathbf{R^2 = 1 - \frac{SSE}{SST}}$$
**SSE = T·ªïng l·ªói b√¨nh ph∆∞∆°ng**
**SST = T·ªïng l·ªói b√¨nh ph∆∞∆°ng**
$$\mathbf{SSE = \sum_{i=1}^{m}(\hat{y}_i - y_i)^2}$$
$$\mathbf{SST = \sum_{i=1}^{m}(y_i - \bar{y}_i)^2}$$
·ªû ƒë√¢y $\mathbf{\hat{y}}$ l√† gi√° tr·ªã d·ª± ƒëo√°n v√† $\mathbf{\bar{y}}$ l√† gi√° tr·ªã trung b√¨nh c·ªßa $\mathbf{y}$.

```python
# Normal equation
y_pred_norm =  np.matmul(X_test_0,theta)


#Evaluvation: MSE
J_mse = np.sum((y_pred_norm - y_test)**2)/ X_test_0.shape[0]


# R_square 
sse = np.sum((y_pred_norm - y_test)**2)
sst = np.sum((y_test - y_test.mean())**2)
R_square = 1 - (sse/sst)
print('The Mean Square Error(MSE) or J(theta) is: ',J_mse)
print('R square obtain for normal equation method is :',R_square)
```
